# Load cell 3: calibration, signal, mean drift force, amplitude, and Grue/Biberg comparison

Denne notatboken er laget for at sensor enkelt skal kunne kjøre analysen.

## Mappestruktur

Legg data i en egen mappe ved siden av denne notatboken:

```text
load_cell_3_sensor_ready.ipynb

    wave_data/

        data6010.dat
        data6015.dat
        data6020.dat
        data605.dat

```

## Hva notatboken gjør

1. Kalibrerer load cell 3.
2. Leser det kalibrerte signalet $F(t)$.
3. Beregner middel driftkraft $F_{\mathrm{drift}}$.
4. Beregner amplitude $A_F$.
5. Sammenligner resultatene med Grue og Biberg (1993).

Notasjonen er holdt i samme stil som i masteroppgaven.


````python
from pathlib import Path
import re

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display

FS = 250.0
g = 9.81
rho_w = 1000.0
r_c = 0.041
LOAD_DIRECTION_SIGN = 1.0

NOTEBOOK_DIR = Path.cwd().resolve()
DATA_DIR = NOTEBOOK_DIR / "sensor_data_load_cell_3"
CALIBRATION_DIR = DATA_DIR / "calibration"
WAVE_DATA_DIR = DATA_DIR / "wave_data"
OUTPUT_DIR = DATA_DIR / "results"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# La listen stå tom for å analysere alle .dat-filer i wave_data.
SELECTED_WAVE_FILES = []

CALIBRATION_FILES = [
    (0, "0g_s.dat"),
    (10, "10g_s.dat"),
    (24, "24g_s.dat"),
    (34, "34g_s.dat"),
]

TIME_WINDOWS_S = [
    {"name": "window_1", "start_s": 0.0, "stop_s": 30.0},
    {"name": "window_2", "start_s": 40.0, "stop_s": 50.0},
    {"name": "window_3", "start_s": 120.0, "stop_s": 150.0},
    {"name": "window_4", "start_s": 170.0, "stop_s": 213.0},
    {"name": "window_5", "start_s": 230.0, "stop_s": 275.0},
    {"name": "window_6", "start_s": 300.0, "stop_s": 335.0},
    {"name": "window_7", "start_s": 350.0, "stop_s": 395.0},
    {"name": "window_8", "start_s": 410.0, "stop_s": 455.0},
    {"name": "window_9", "start_s": 470.0, "stop_s": 515.0},
    {"name": "window_10", "start_s": 530.0, "stop_s": 575.0},
]

WAVE_WINDOW_PAIRS = [
    {"wave_label": "60", "reference_idx": 0, "wave_idx": 1},
    {"wave_label": "66", "reference_idx": 2, "wave_idx": 3},
    {"wave_label": "70", "reference_idx": 4, "wave_idx": 5},
    {"wave_label": "76", "reference_idx": 6, "wave_idx": 7},
    {"wave_label": "80", "reference_idx": 8, "wave_idx": 9},
]

DISPLAY_DISTANCE_MAP_CM = {
    20: 22.5,
    15: 17.5,
    10: 12.5,
    5: 7.5,
    0: 0.0,
}

WAVE_METRICS = {
    "60": {"a_m": 0.600 / 100.0, "k_1_per_m": 15.679},
    "66": {"a_m": 0.6552 / 100.0, "k_1_per_m": 18.3010},
    "70": {"a_m": 0.692 / 100.0, "k_1_per_m": 20.049},
    "76": {"a_m": 0.7820 / 100.0, "k_1_per_m": 23.6334},
    "80": {"a_m": 0.842 / 100.0, "k_1_per_m": 26.023},
}

GRUE_BIBERG_1993_FIG7A_DOTTED = np.array(
    [
        [0.20, 0.00],
        [0.40, 0.09],
        [0.60, 0.25],
        [0.80, 0.46],
        [1.00, 0.57],
        [1.10, 0.60],
        [1.25, 0.57],
    ],
    dtype=float,
)

DATA_WAVE_FILE_PATTERN = re.compile(
    r"^da?ata(?P<wave>[4-9]\d)(?P<distance>\d*)(?P<model>[cks])?(?P<run>\d*)\.dat$",
    re.IGNORECASE,
)

print(f"Notebook directory: {NOTEBOOK_DIR}")
print(f"Data directory:     {DATA_DIR}")
print(f"Output directory:   {OUTPUT_DIR}")
````

````python
def wl_col_name(df):
    preferred_columns = ["WL4_2", "WL2_2", "WL1_2", "WL4_1", "WL2_1", "WL1_1"]
    for column in preferred_columns:
        if column in df.columns:
            return column
    for column in df.columns:
        if re.match(r"^WL\d+_\d+$", str(column)):
            return str(column)
    raise KeyError(f"Fant ingen WL-kolonne. Kolonner: {list(df.columns)}")


def read_wl_file(path):
    df = pd.read_csv(path, sep="\t", decimal=",", low_memory=False)
    wl_column = wl_col_name(df)
    values = df[wl_column].astype(str).str.strip()
    values = values.str.replace(".", "", regex=False).str.replace(",", ".", regex=False)
    wl = pd.to_numeric(values, errors="coerce").dropna().reset_index(drop=True).to_numpy(dtype=float)
    if "msec" in df.columns:
        t = pd.to_numeric(df["msec"], errors="coerce").to_numpy(dtype=float)[: len(wl)] / 1000.0
    else:
        t = np.arange(len(wl), dtype=float) / FS
    return t, wl


def parse_wave_file_name(filename):
    match = DATA_WAVE_FILE_PATTERN.match(Path(filename).name)
    if not match:
        raise ValueError(f"Kunne ikke tolke filnavn: {filename}")
    distance_code = match.group("distance") or "0"
    return {
        "source_wave": int(match.group("wave")),
        "distance_code_cm": int(distance_code),
    }


def selected_wave_paths():
    if not WAVE_DATA_DIR.exists():
        raise FileNotFoundError(f"Could not find the wave_data folder: {WAVE_DATA_DIR}")
    if SELECTED_WAVE_FILES:
        paths = [WAVE_DATA_DIR / name for name in SELECTED_WAVE_FILES]
        missing = [path for path in paths if not path.exists()]
        if missing:
            raise FileNotFoundError("Mangler bølgefiler:\n" + "\n".join(str(path) for path in missing))
        return paths
    paths = sorted(WAVE_DATA_DIR.glob("*.dat"))
    if not paths:
        raise FileNotFoundError(f"Fant ingen .dat-filer i {WAVE_DATA_DIR}")
    return paths


def half_peak_to_peak(signal):
    values = np.asarray(signal, dtype=float)
    values = values[np.isfinite(values)]
    if values.size == 0:
        return np.nan
    return 0.5 * (np.nanmax(values) - np.nanmin(values))


def window_values(t, signal, window):
    mask = (t >= window["start_s"]) & (t <= window["stop_s"])
    values = np.asarray(signal[mask], dtype=float)
    return values[np.isfinite(values)]


def wl_to_force(wl, C_cal, b_cal, WL_0):
    return C_cal * (np.asarray(wl, dtype=float) - WL_0) + b_cal


def k_r_c_value(wave_label):
    return WAVE_METRICS[wave_label]["k_1_per_m"] * r_c


def normalized_drift_force(F_drift, wave_label):
    a = WAVE_METRICS[wave_label]["a_m"]
    return F_drift / (rho_w * g * a**2 * r_c)
````

## 1. Kalibrering

This cell computes the calibration curve.

$$
F = C_{\mathrm{cal}}(WL - WL_0) + b_{\mathrm{cal}}.
$$


````python
calibration_rows = []

for mass_g, filename in CALIBRATION_FILES:
    path = CALIBRATION_DIR / filename
    if not path.exists():
        raise FileNotFoundError(f"Missing calibration file: {path}")
    t_cal, wl_cal = read_wl_file(path)
    calibration_rows.append(
        {
            "file": filename,
            "m_g": mass_g,
            "F_N": mass_g / 1000.0 * g,
            "WL_mean": np.nanmean(wl_cal),
            "WL_std": np.nanstd(wl_cal),
            "duration_s": np.nanmax(t_cal),
        }
    )

calibration_df = pd.DataFrame(calibration_rows).sort_values("m_g").reset_index(drop=True)
WL_0 = calibration_df.loc[calibration_df["m_g"] == 0, "WL_mean"].iloc[0]
calibration_df["dWL"] = calibration_df["WL_mean"] - WL_0
C_cal, b_cal = np.polyfit(calibration_df["dWL"], calibration_df["F_N"], 1)
calibration_df["F_fit_N"] = C_cal * calibration_df["dWL"] + b_cal
calibration_df["fit_error_N"] = calibration_df["F_N"] - calibration_df["F_fit_N"]

print(f"Calibration curve: F = {C_cal:.12g} * (WL - WL_0) + {b_cal:.12g}")
print(f"WL_0 = {WL_0:.6f}")

display(calibration_df.round(6))
calibration_df.to_csv(OUTPUT_DIR / "calibration_results.csv", index=False, encoding="utf-8-sig")

fig, axes = plt.subplots(2, 1, figsize=(10, 8))
axes[0].errorbar(calibration_df["m_g"], calibration_df["WL_mean"], yerr=calibration_df["WL_std"], fmt="o", capsize=4)
axes[0].set_title("Load cell 3 calibration")
axes[0].set_xlabel(r"$m$ [g]")
axes[0].set_ylabel(r"$WL$")
axes[0].grid(True, alpha=0.3)

x_fit = np.linspace(calibration_df["dWL"].min(), calibration_df["dWL"].max(), 200)
axes[1].scatter(calibration_df["dWL"], calibration_df["F_N"], s=70, label="Målepunkter")
axes[1].plot(x_fit, C_cal * x_fit + b_cal, color="black", linewidth=2, label="Lineær tilpasning")
axes[1].set_xlabel(r"$\Delta WL = WL - WL_0$")
axes[1].set_ylabel(r"$F$ [N]")
axes[1].grid(True, alpha=0.3)
axes[1].legend()
plt.tight_layout()
plt.show()
````

## 2. Signal, middel driftkraft og amplitude

For hvert bølgetilfelle beregnes:

- referansemiddel $F_{\mathrm{ref}}$
- bølgemiddel $F_{\mathrm{wave}}$
- middel driftkraft $F_{\mathrm{drift}} = F_{\mathrm{wave}} - F_{\mathrm{ref}}$
- amplitude $A_F = \frac{1}{2}(F_{\max} - F_{\min})$


````python
result_rows = []

for signal_path in selected_wave_paths():
    meta = parse_wave_file_name(signal_path.name)
    t, wl = read_wl_file(signal_path)
    F_t = LOAD_DIRECTION_SIGN * wl_to_force(wl, C_cal=C_cal, b_cal=b_cal, WL_0=WL_0)
    d_cm = DISPLAY_DISTANCE_MAP_CM.get(meta["distance_code_cm"], float(meta["distance_code_cm"]))

    case_rows = []
    for pair in WAVE_WINDOW_PAIRS:
        wave_label = pair["wave_label"]
        ref_window = TIME_WINDOWS_S[pair["reference_idx"]]
        wave_window = TIME_WINDOWS_S[pair["wave_idx"]]

        F_ref = window_values(t, F_t, ref_window)
        F_wave = window_values(t, F_t, wave_window)

        F_ref_mean = np.nanmean(F_ref) if F_ref.size else np.nan
        F_wave_mean = np.nanmean(F_wave) if F_wave.size else np.nan
        F_drift = F_wave_mean - F_ref_mean
        A_F = half_peak_to_peak(F_wave)

        row = {
            "signal_file": signal_path.name,
            "source_wave": meta["source_wave"],
            "wave_label": wave_label,
            "d_cm": d_cm,
            "F_ref_mean_N": F_ref_mean,
            "F_wave_mean_N": F_wave_mean,
            "F_drift_N": F_drift,
            "A_F_N": A_F,
            "k_r_c": k_r_c_value(wave_label),
            "F_drift_over_rho_w_g_a2_r_c": normalized_drift_force(F_drift, wave_label),
        }
        result_rows.append(row)
        case_rows.append(row)

    case_df = pd.DataFrame(case_rows)
    display(case_df.round(6))

    fig, ax = plt.subplots(figsize=(14, 6))
    ax.plot(t, F_t, color="#ff7f0e", linewidth=1.1, label=r"$F(t)$")

    for pair in WAVE_WINDOW_PAIRS:
        wave_label = pair["wave_label"]
        ref_window = TIME_WINDOWS_S[pair["reference_idx"]]
        wave_window = TIME_WINDOWS_S[pair["wave_idx"]]
        row = case_df.loc[case_df["wave_label"] == wave_label].iloc[0]

        ax.axvspan(ref_window["start_s"], ref_window["stop_s"], color="#c6dbef", alpha=0.25)
        ax.axvspan(wave_window["start_s"], wave_window["stop_s"], color="#fdd0a2", alpha=0.25)
        ax.hlines(row["F_ref_mean_N"], ref_window["start_s"], ref_window["stop_s"], colors="#08519c", linewidth=2.2)
        ax.hlines(row["F_wave_mean_N"], wave_window["start_s"], wave_window["stop_s"], colors="#a63603", linewidth=2.2, linestyles="--")

    ax.set_title(f"Kalibrert kraftsignal: {signal_path.name}")
    ax.set_xlabel(r"$t$ [s]")
    ax.set_ylabel(r"$F(t)$ [N]")
    ax.grid(True, alpha=0.3)
    ax.legend(framealpha=0.9, loc="upper left")
    plt.tight_layout()
    plt.show()
    fig.savefig(OUTPUT_DIR / f"{signal_path.stem}_signal.png", dpi=300, bbox_inches="tight")

results_df = pd.DataFrame(result_rows)
results_df.to_csv(OUTPUT_DIR / "load_cell_3_results.csv", index=False, encoding="utf-8-sig")
print(f"Saved: {OUTPUT_DIR / 'load_cell_3_results.csv'}")
````

## 3. Sammenligning med Grue og Biberg (1993)

Den dimensjonsløse størrelsen som plottes er

$$
\frac{F_{\mathrm{drift}}}{\rho_w g a^2 r_c}
$$

som funksjon av

$$
k r_c.
$$


````python
plot_df = results_df[results_df["d_cm"] > 0].copy()

mean_curve_df = (
    plot_df.groupby(["wave_label", "k_r_c"], as_index=False)["F_drift_over_rho_w_g_a2_r_c"]
    .mean()
    .sort_values("k_r_c")
)

display(mean_curve_df.round(6))

fig, ax = plt.subplots(figsize=(8.5, 6))
ax.plot(
    GRUE_BIBERG_1993_FIG7A_DOTTED[:, 0],
    GRUE_BIBERG_1993_FIG7A_DOTTED[:, 1],
    linestyle=":",
    linewidth=2.8,
    color="black",
    marker="o",
    label="Grue and Biberg (1993), Fig. 7a",
)

for distance_cm, group in plot_df.groupby("d_cm"):
    ax.scatter(
        group["k_r_c"],
        group["F_drift_over_rho_w_g_a2_r_c"],
        s=70,
        alpha=0.9,
        label=f"d = {distance_cm:g} cm",
    )

ax.plot(
    mean_curve_df["k_r_c"],
    mean_curve_df["F_drift_over_rho_w_g_a2_r_c"],
    linestyle="--",
    linewidth=2.6,
    color="#b22222",
    marker="o",
    label="Middelkurve fra load cell 3",
)

ax.set_title("Sammenligning med Grue og Biberg")
ax.set_xlabel(r"$k r_c$ [-]")
ax.set_ylabel(r"$F_{\mathrm{drift}} / (\rho_w g a^2 r_c)$ [-]")
ax.set_xlim(0.2, 1.2)
ax.grid(True, alpha=0.3)
ax.legend(framealpha=0.9, loc="upper left")
plt.tight_layout()
plt.show()

fig.savefig(OUTPUT_DIR / "grue_biberg_comparison.png", dpi=300, bbox_inches="tight")
mean_curve_df.to_csv(OUTPUT_DIR / "mean_curve_results.csv", index=False, encoding="utf-8-sig")
print(f"Saved: {OUTPUT_DIR / 'grue_biberg_comparison.png'}")
print(f"Saved: {OUTPUT_DIR / 'mean_curve_results.csv'}")
````
